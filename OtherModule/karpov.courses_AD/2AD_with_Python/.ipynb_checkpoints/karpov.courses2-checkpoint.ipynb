```python
df = df.rename(columns={'Номер':'number',
               'Дата создания':'create_date',
               'Дата оплаты':'payment_date',
               'Title':'title',
               'Статус':'status',
               'Заработано':'money',
               'Город':'city',
               'Платежная система':'payment_system'})
```

```python
all_money = df.money.sum()
all_money
```

```python
df.groupby('title', as_index=False) \
.aggregate({'money': 'sum'}) \
.sort_values('money', ascending=False)
```

```python
df.groupby(['title', 'city'], as_index=False) \
.aggregate({'money': 'sum'}) \
.sort_values('money', ascending=False)
```

```python
money_by_city = df.groupby(['title', 'city'], as_index=False) \
.aggregate({'money': 'sum'}) \
.sort_values('money', ascending=False)
```

```python
money_by_city.to_csv('money_by_city.csv', index=False)
```

```python
money_title = df \
.query("status == 'Завершен'") \
.groupby('title', as_index=False) \
.aggregate({'money': 'sum', 'number': 'count'}) \
.sort_values('money', ascending=False) \
.rename(columns={'number': 'success_orders'})

money_title
```

```python
today_day = datetime.today().strftime('%Y-%m-%d')
file_name = 'money_title_{}.csv'
file_name = file_name.format(today_day)
```

```python
count_title == len(money_title)
```

```python
if int(money_title.money.sum()) == int(all_money):
    money_title.to_csv(file_name, index=False)
    print('Success! File {} wrote.'.format(file_name))
else:
    print('DATA Error!')
```

```python
taxi = pd.read_csv('https://stepik.org/media/attachments/lesson/360340/2_taxi_nyc.csv')
taxi.shape
```

```python
taxi
```

```python
taxi.dtypes
```

```python
taxi["borough"].value_counts()
```

```python
taxi["pickups"].sum()
```

```python
taxi.groupby('borough', as_index=False).aggregate({'pickups': 'sum'}).sort_values('pickups', ascending=False)
```

```python
min_pickups = taxi.groupby('borough').aggregate({'pickups': 'sum'}).idxmin()
min_pickups
```

```python
taxi.groupby(['borough', 'hday'], as_index=False).aggregate({'pickups': 'mean'}).sort_values('pickups', ascending=False)
```

```python
pickups_by_mon_bor = taxi[['pickup_month', 'borough', 'pickups']].groupby(['borough', 'pickup_month'], as_index=False).aggregate({'pickups': 'sum'}).sort_values('pickups', ascending=False)
pickups_by_mon_bor
```

```python
def temp_to_celcius(inp_temp):            
    return ((inp_temp-32)*5/9)
```

```python
taxi['temp_C'] = temp_to_celcius(taxi['temp'])
```

```python
import pandas as pd
bookings = pd.read_csv('https://stepik.org/media/attachments/lesson/360344/bookings.csv', sep=';')
bookings_head = bookings.head(7)
bookings_head
```

```python
bookings.shape
```

```python
bookings.info()
```

```python
bookings.dtypes
```

```python
bookings.columns
```

```python
bookings.rename(str.lower, axis=1, inplace=True) #
```

```python
for i in bookings.columns:
    bookings.rename(columns={i:i.replace(' ','_')}, inplace=True)
    
```

```python
bookings.columns
```

```python
bookings \
.query("is_canceled == 0") \
.groupby('country', as_index=False) \
.aggregate({'is_canceled': 'count'}) \
.sort_values('is_canceled', ascending=False)
```

```python
bookings \
.query("hotel == 'City Hotel'") \
.aggregate({'stays_total_nights': 'mean'}).round(2)
```

```python
bookings \
.query("hotel == 'Resort Hotel'") \
.aggregate({'stays_total_nights': 'mean'}).round(2)
```

```python
bookings \
.query("reserved_room_type != assigned_room_type").shape
```

```python
bookings \
.query("is_canceled == 1").shape
```

```python
resv_count1 = bookings \
.query("reserved_room_type != assigned_room_type & is_canceled == 0") \
.shape[0]
```

```python
resv_count2 = bookings \
.query("is_canceled == 1") \
.shape[0]
```

```python
resv_count1 + resv_count2
```

```python
bookings \
.groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
.aggregate({'reservation_status': 'count'}) \
.sort_values(['arrival_date_year','reservation_status'], ascending=[True, False])
```

```python
bookings \
.query("arrival_date_year == 2016") \
.arrival_date_month.value_counts()
```

```python
bookings \
.query("arrival_date_year == 2017") \
.arrival_date_month.value_counts()
```

```python
bookings \
.query("hotel == 'City Hotel' & is_canceled ==1") \
.groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
.aggregate({'is_canceled': 'count'}) \
.sort_values(['arrival_date_year','is_canceled'], ascending=[True, False])
```
