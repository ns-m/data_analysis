```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Product Gallery.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Year'] = df['OrderDate'].dt.year
df['Month'] = df['OrderDate'].dt.month
df['Day'] = df['OrderDate'].dt.day
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_veget = df[df['Category'] == 'Приправы']
df_veget.head()
```

```python
visual = df_veget[['OrderDate', 'Sales', 'Profit', 'Quantity']]
```

```python
visual
```

```python
visual.plot(x ='OrderDate', y = ['Sales', 'Profit'], figsize=(12, 6))
```

```python
sns.set_theme(style='ticks')
sns.lineplot(visual, x ='OrderDate', y = 'Sales')
```

```python
visual2 = df_veget[['Sales', 'Profit']]
```

```python
visual2.plot.hist()
```

```python
sns.distplot(visual2)
```

```python
sns.heatmap(visual2.corr(), annot=True)
```

```python
sns.relplot(
    data = df,
    x = 'Sales', y = 'Profit',
    hue = 'Year', size = 'Quantity',
    sizes = (10, 200)
)
```
