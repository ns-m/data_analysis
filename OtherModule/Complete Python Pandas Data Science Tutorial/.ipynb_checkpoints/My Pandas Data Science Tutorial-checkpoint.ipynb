## Aggregate Statistics (Groupby)

```python
poke_df_csv.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)
```

```python
poke_df_csv.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)
```

```python
poke_df_csv['count'] = 1
poke_df_csv.groupby(['Type 1']).count()['count'].sort_values(ascending=False)
```

```python
poke_df_csv.groupby(['Type 1', 'Type 2']).count()['count'].sort_values(ascending=False)
```

## Working with large amounts of data

```python
for df in pd.read_csv('modi_poke.csv', chunksize=10):
    print('---PART DF---')
    print(df)
```

```python
new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv('modi_poke.csv', chunksize=10):
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, results])
    
new_df    
```
