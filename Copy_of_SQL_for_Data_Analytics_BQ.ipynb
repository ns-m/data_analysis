##<div align="left" id="agg"><a href="#contents"><font size="4" color="red"><b>AGGREGATION</b></font></a></div>

SQL предоставляет несколько способов агрегирования информации на разных уровнях. В конце вам нужно будет использовать «GROUP BY» и указать, в каких столбцах вы собираете данные. <br>
Вот некоторые примеры -

```python
%%bigquery --project $PROJECT_ID

SELECT
    df.Embarked, df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets
FROM ML_BQ.train_Titanic AS df
GROUP BY df.Embarked, df.Sex
```

Помимо агрегатных функций, использованных выше, многие диалекты SQL также поддерживают STDEV для стандартного отклонения, VAR для дисперсии, SQRT для квадратного корня и многие другие. Вы всегда можете воспользоваться Google, чтобы узнать, какие агрегатные функции поддерживаются используемым вами диалектом SQL.

```python
#Aggregating information at "Embarked" and "Sex" levels
%%bigquery --project $PROJECT_ID

SELECT
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets

FROM ML_BQ.train_Titanic AS df
GROUP BY df.Embarked,
    df.Sex
```

"WHERE" всегда используется перед "GROUP BY" <br>
Вы также можете отсортировать вывод, используя «ORDER BY» в конце

```python
#Aggregating information at "Embarked" and "Sex" levels
#and also using "WHERE" to filter basis some conditions
%%bigquery --project $PROJECT_ID

SELECT
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets
FROM ML_BQ.train_Titanic AS df

WHERE age >= 18
    AND age <=60

GROUP BY df.Embarked,
    df.Sex
```

По умолчанию сортировка выполняется в порядке возрастания, но вы можете изменить его на порядок убывания, используя ключевое слово «DESC» в операторе «ORDER BY».

```python
#Aggregating information at "Embarked" and "Sex" levels
#and also using "WHERE" to filter basis some conditions
#and also sorting the output
%%bigquery --project $PROJECT_ID

SELECT
    df.Sex,df.Embarked,

    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets

FROM ML_BQ.train_Titanic AS df

WHERE age >= 18
    AND age <=60

GROUP BY df.Embarked,
    df.Sex

ORDER BY  df.Sex, df.Embarked Desc
```

<b> HAVING </b><br>
Lastly, once you have aggregated the information, you can further put filters on the aggregated output using "HAVING". Always remember the difference between "WHERE" and "HAVING" as it is one of the most commonly asked interview questions. "WHERE" is used to filter the data directly from the tables while "HAVING" is used to filter the aggregated data you get as the output of a SQL

```python
#Aggregating information at "Embarked" and "Sex" levels
#and also using "WHERE" to filter basis some conditions
#and also sorting the output in descending order
%%bigquery --project $PROJECT_ID

SELECT
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets

FROM ML_BQ.train_Titanic AS df

WHERE age >= 18
    AND age <=60

GROUP BY df.Embarked,
    df.Sex

ORDER BY df.Embarked DESC,
    df.Sex DESC
```

<b> HAVING </b><br>
Наконец, после того, как вы обобщили информацию, вы можете добавить фильтры к агрегированному выводу, используя «HAVING». Всегда помните разницу между "WHERE" и "HAVING" , поскольку это один из наиболее часто задаваемых вопросов на собеседовании. «WHERE» используется для фильтрации данных непосредственно из таблиц, в то время как «HAVING» используется для фильтрации агрегированных данных, которые вы получаете в качестве вывода SQL.

```python
#Aggregating information at "Embarked" and "Sex" levels
#and also using "WHERE" to filter basis some conditions
#and also using "HAVING" for fitering the output for buckets where average age is more than 30
#and also using "ORDER BY" for sorting the output in descending order
%%bigquery --project $PROJECT_ID

SELECT
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets

FROM ML_BQ.train_Titanic AS df

WHERE age >= 18
    AND age <=60

GROUP BY df.Embarked,
    df.Sex

HAVING avg_age > 30

ORDER BY df.Embarked DESC,
    df.Sex DESC
```

<b>Помните порядок. Сначала мы использовали "WHERE", потом "GROUP BY", потом "HAVING" и после "ORDER BY"</b>

Перед объединением любых двух таблиц необходимо знать уровни/столбцы, на которых данные уникальны в обеих таблицах. Например, набор данных titanic уникален на уровне столбца «PassengerId», а данные городов уникальны на уровне столбца «код». Непонимание уникальности данных является основной причиной неправильных результатов в SQL. Помня об уровнях данных, вы обеспечите значимые соединения.

<b>INNER JOIN</b><br>
INNER JOIN дает строки, соответствующие только общим значениям (в столбцах, используемых в условии внутреннего соединения) между двумя таблицами.
<img src = "https://www.w3schools.com/sql/img_innerjoin.gif" width="25%" height="25%"><br>

```python
query = """

SELECT
    c.city,
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets

FROM df

INNER JOIN cities AS c
    ON df.Embarked = c.code

GROUP BY df.Embarked,
    df.Sex
"""

sqldf.run(query)
```

Как видите, с помощью INNER JOIN мы получаем только общие элементы - "S" и "C", но не "L" (присутствует только в "cities") и "Q" (присутствует только в "titanic_disaster_dataset")

<b>LEFT JOIN</b><br>

LEFT JOIN дает все строки из левой таблицы и только те строки из правой таблицы, где значения являются общими в столбцах, используемых в условии левого соединения.
<img src = "https://www.w3schools.com/sql/img_leftjoin.gif" width="25%" height="25%"><br>

```python
query = """

SELECT
    c.city,
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets

FROM df

LEFT JOIN cities AS c
    ON df.Embarked = c.code

GROUP BY df.Embarked,
    df.Sex

"""
sqldf.run(query)
```

Как видите, с помощью LEFT JOIN мы получаем все строки из левой таблицы ("S", "Q", "C" - все присутствуют в выводе), но только для "C" и "S", мы получают "город" из правой таблицы

<b>В основном вы будете использовать INNER и LEFT JOIN в своей карьере, но полезно знать и о других соединениях<b>

<b>RIGHT JOIN</b><br>
RIGHT JOIN является зеркальной противоположностью LEFT JOIN. ПРАВОЕ СОЕДИНЕНИЕ дает все строки из правой таблицы и только те строки из левой таблицы, где значения являются общими в столбцах, используемых в условии правого соединения.
<img src = "https://www.w3schools.com/sql/img_rightjoin.gif" width="25%" height="25%"><br>

К сожалению, «ПРАВОЕ СОЕДИНЕНИЕ» не поддерживается sqldf. Но мы можем увидеть результат RIGHT JOIN, поменяв местами таблицы в запросе LEFT JOIN выше. Вот как это будет выглядеть

```python
query = """

SELECT
    c.city,
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets

FROM cities AS c

LEFT JOIN df
    ON df.Embarked = c.code

GROUP BY df.Embarked,
    df.Sex

"""
sqldf.run(query)
```

<div align="left" id ="adv_union"> <a href="#adv"><font size="3" color="black"><b>Объединить и объединить все</b></font> </a></div>
Если у вас есть два набора данных с одинаковыми столбцами, и вы хотите наложить один поверх другого и получить все отдельные строки, вы можете использовать оператор UNION. Каждый оператор SELECT в рамках UNION должен иметь одинаковое количество полей в наборах результатов с одинаковыми типами данных, как показано ниже.

```python
#Using "UNION" to stack datasets
query = """

SELECT DISTINCT df.Embarked FROM df
UNION
SELECT DISTINCT c.code FROM cities AS c

"""
sqldf.run(query)
```

"UNION" дает вам только отдельные строки после укладки. Если вам не нужны отдельные строки, вы можете использовать  "UNION ALL", которое просто сложит ( stack) две таблицы друг над другом.

```python
#Using "UNION ALL" to stack datasets
query = """

SELECT DISTINCT df.Embarked FROM df
UNION ALL
SELECT DISTINCT c.code FROM cities AS c

"""
sqldf.run(query)
```

##<div align="left" id ="adv_rank"> <a href="#adv"><font size="3" color="black"><b>Ranking in SQL</b></font></a></div>
Часто бывает, что вам нужно отсортировать данные и присвоить ранг каждой строке в наборе данных. RANK(), DENSE_RANK() и ROW_NUMBERS() — три функции, которые помогут вам сделать именно это. Обычно эти функции используются с функцией OVER() для сортировки данных на основе некоторого столбца перед присвоением ранга каждой строке, а также для разделения данных, о чем мы узнаем позже. Сначала рассмотрим RANK().<br><br>
RANK() предоставляет ранг каждой строке. Если любые n строк имеют одинаковое значение в столбце сортировки, то RANK() дает одинаковое значение всем этим строкам. Для следующей строки, где значение отличается от предыдущих n строк, присвоенный ранг будет равен предыдущему рангу плюс n. Вот пример -

```python
#RANK() sorts the data, gives the same rank to all consecutive n rows with same values,
#but for next row with a different value, it gives a rank equal to previous rank plus n
%%bigquery --project $PROJECT_ID

SELECT
    df.Name,
    --df.Embarked,
    --df.Sex,
    df.Age,
    df.Fare,
    RANK() OVER(ORDER BY df.Age DESC) AS rnk
FROM ML_BQ.train_Titanic AS df
order by rnk
limit 20
```

Как видите, «OVER(ORDER BY df.Age DESC)» упорядочивает возраст в порядке убывания, а затем используется Rank() для присвоения ранга каждой строке. Обратите внимание на строки №2 и №3 (3 и 4), где значения возраста одинаковы (например, 71), и, следовательно, обе строки имеют одинаковый ранг (= 3). В строке 5 ранжирование начинается с 5 (=предыдущий ранг 3 плюс 2), а не с 4.<br><br>

Таким образом, при использовании RANK() некоторые ранги пропускаются, если в порядке сортировки совпадают значения. Используя DENSE_RANK(), мы можем гарантировать, что ни один ранг не будет пропущен. Обратите внимание на пример ниже, где строка 5 имеет ранг 4, а не 5.

```python
#DENSE_RANK() does not skip ranks like RANK() does
%%bigquery --project $PROJECT_ID

SELECT
    df.Name,
    --df.Embarked,
    --df.Sex,
    df.Age,
    df.Fare,
    DENSE_RANK() OVER(ORDER BY df.Age DESC) AS rnk

FROM ML_BQ.train_Titanic AS df
order by rnk
limit 20
```

DENSE_RANK() — наиболее часто используемая функция из этих трех. Но в зависимости от вашего варианта использования вы можете выбрать любую из трех вышеперечисленных функций.

До сих пор мы использовали только один столбец для сортировки данных, мы также можем использовать несколько столбцов для сортировки данных. Мы также можем указать, в каком порядке мы хотим сортировать данные. Вот пример, в котором мы сначала сортируем данные по возрасту, а если возраст тот же, то сортируем по тарифу:

```python
#ROW_NUMBER() sorts the data and gives different ranks to each row
%%bigquery --project $PROJECT_ID

SELECT
    df.Name,
    df.Embarked,
    df.Sex,
    df.Age,
    df.Fare,
    ROW_NUMBER() OVER(ORDER BY df.Age DESC) AS rnk

FROM ML_BQ.train_Titanic AS df
```

Кроме того, OVER() также поддерживает разделение данных. Предположим, вы хотите дать отдельные ранги для определенных уровней данных, тогда вы можете использовать "PARTITION BY" («РАЗДЕЛЕНИЕ ПО») в функции OVER(). В приведенном ниже примере для каждой комбинации df.Embarked и df.Sex мы получим разные наборы рангов.

```python
query = """

SELECT
    df.Name,
    df.Embarked,
    df.Sex,
    df.Age,
    df.Fare,
    RANK() OVER(PARTITION BY df.Embarked, df.Sex
                ORDER BY df.Age DESC, df.Fare DESC) AS rnk

FROM df

"""
sqldf.run(query)
```

<div align="left" id ="adv_ll"> <a href="#adv"><font size="3" color="black"><b>Lead and Lag</b></font></a></div>
<div align="left" id ="adv_ll"> <a href="#adv"><font size="3" color="black"><b>Опережение и отставание</b></font></a></div>

LEAD() и LAG() можно использовать для получения **n-го предыдущего или n-го следующего значения** в столбце после сортировки данных с помощью OVER().

```python
#Examples for LEAD() and LAG()
query = """


SELECT
    df.Name,
    LEAD(df.Name, 1) OVER(ORDER BY df.Name) AS next_name,
    LEAD(df.Name, 2) OVER(ORDER BY df.Name) AS second_next_name,
    LAG(df.Name, 1) OVER(ORDER BY df.Name) AS previous_name,
    LAG(df.Name, 2) OVER(ORDER BY df.Name) AS second_previous_name

FROM df
"""
sqldf.run(query)
```

Используя «PARTITION BY» в функции OVER(), вы можете разделить данные и получить LEAD() и LAG() столбца в partitions.

<div align="left" id ="adv_sub"> <a href="#adv"><font size="3" color="black"><b>Subqueries</b></font></a></div>
<div align="left" id ="adv_sub"> <a href="#adv"><font size="3" color="black"><b>Подзапросы</b></font></a ></дел>

Подзапросы — это небольшие запросы, которые используются внутри более крупного основного SQL-запроса. Подзапросы временно хранят выходные данные SQL, и поверх них можно записать дополнительные запросы SQL.<br>

Вот несколько примеров подзапросов:

```python
#Using subqueries to get average age and then filtering only those rows where average age is below 30
%%bigquery --project $PROJECT_ID

SELECT

    df1.Embarked,
    df1.avg_age

FROM (SELECT
            a.Embarked,
            AVG(a.Age) AS avg_age
        FROM ML_BQ.train_Titanic AS a
        GROUP BY a.Embarked) AS df1

WHERE df1.avg_age<30
```

Здесь df1 — это временный набор данных, который создается с помощью подзапроса.

```python
#Using subqueries to get average age and number of passengers separately and then joining the outputs
%%bigquery --project $PROJECT_ID

SELECT
    df1.Embarked,
    df1.avg_age,
    df2.tot_passengers
FROM (SELECT
            a.Embarked,
            AVG(a.Age) AS avg_age
        FROM ML_BQ.train_Titanic AS a
        GROUP BY a.Embarked) AS df1

INNER JOIN (SELECT
                b.Embarked,
                COUNT(b.PassengerID) AS tot_passengers
            FROM ML_BQ.train_Titanic AS b
            GROUP BY b.Embarked) AS df2
    ON df1.Embarked = df2.Embarked
```

Вы можете вкладывать подзапросы в другие подзапросы, а также использовать их для фильтрации набора данных, как показано ниже:

```python
#Using nested subqueries to filter data
%%bigquery --project $PROJECT_ID

SELECT b.* FROM ML_BQ.train_Titanic AS b
WHERE b.Embarked IN
       (SELECT
            DISTINCT df1.Embarked
        FROM (SELECT
                    a.Embarked,
                    AVG(a.Age) AS avg_age
                FROM ML_BQ.train_Titanic AS a
                GROUP BY a.Embarked) AS df1
        WHERE df1.avg_age>30)
```

<div align="left" id ="adv_type"> <a href="#adv"><font size="3" color="black"><b>Приведение типов</b></font></a ></дел>

Вы можете использовать функцию CAST() для изменения типа данных столбцов в наборе данных. Вот несколько примеров -

```python
#Using CAST() to change datatype of columns
query = """

SELECT
    df.Name,
    df.Fare,
    CAST(df.Fare AS INT) AS integer_fare,
    CAST(CAST(df.Fare AS INT) AS FLOAT) AS float_fare
FROM df
order by integer_fare desc
LIMIT 10

"""
sqldf.run(query)
```
